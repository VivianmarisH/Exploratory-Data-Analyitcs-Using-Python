# Exploratory Data Aalytics

```python
df["Unit_Price"] = round(df["Total"] / df["Qty Ordered"], 2)
df.head()
```

```python
df = df.rename(columns={'Total': 'Revenue'})
df.head()
```

```python
# Split the 'Full Name' column into 'First Name' and 'Last Name' columns
df[['First Name', 'Last Name']] = df['Full Name'].str.split(' ', n=1, expand=True)

plt.show()
```

```python
#Extrat the month, quater and Year from Date_of_Order column
df['Date of Order'] = pd.to_datetime(df['Date of Order'])

#df["month"] = df["Date_of_Order"].dt.month
df['year'] = df['Date of Order'].dt.year
df['month'] = df['Date of Order'].dt.month
df['month_name'] = df['Date of Order'].dt.month_name()
df['quarter'] = df['Date of Order'].dt.quarter

df.head(3)
```

```python
# Create pivot table with years as columns
pivot_df = pd.pivot_table(df, values='Revenue', index=df.index, columns='year')

# Add the years as columns to the table
years = df['year'].unique()
for year in years:
    pivot_df[f'year_{year}'] = pivot_df[year]
    
# Drop the original year columns
pivot_df.drop(columns=years, inplace=True)
    
# Display the table
print(pivot_df)
```

```python
#Creating a function for percentage calculation and application on plots    
def perc_plot(plot, feature):
    total = len(feature)
    for p in plot.patches:
        perc = '{:.1f}%'.format(100 * p.get_height() / total)  # Fixed formatting issue here
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_y() + p.get_height()
        plot.annotate(perc, (x, y), ha='center')
    plt.show()
```
